First, let's start by importing everything we need!

```python
import json
import time
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
from collections import namedtuple
from ipywidgets import widgets  
from IPython.display import display
% matplotlib inline
```

This is the path to the json file containing the cleaned tweets. We can't push this to github, because Twitter's API terms of service do not allow this

```python
filename = './data/cleanedShootingTweets.json'
```

This is the read and aggregate function. This is a first pass at creating a function that reads through the data bit by bit, and aggregates. This function takes in the number of lines to read at a time and a filepath, and returns a dictionary of hashtags and counts of those hashtags. 

```python
#This works if we don't care about time stuff. 
def ReadAndAggregate(linesAtATime,filepath):
    i = 0
    data = []
    hashtagsDict = {}
    
    #Open the file     
    with open(filename) as cleanedTweets:
        # Loop through all the tweets        
        for tweet in cleanedTweets:
            i += 1
            jsonline = json.loads(tweet)
            data.append(jsonline)
            #aggregate after we've read through the right number of lines
            if (i % linesAtATime == 0):
                print i
                df = pd.DataFrame(data=data)
                allHashtags = df.entities_hashtags_text.tolist()
                #                 
                for tweet in allHashtags:
                    for hashtag in tweet:
                        hashtagsDict[hashtag.upper()] = hashtagsDict.get(hashtag.upper(), 0) + 1
                
                data = []
    return hashtagsDict
```

Now, let's try out this function!

```python
hashtagsDict = ReadAndAggregate(100000, filename)
```

Now, we want to actually plot the top hashtags. First we need to get the hashtags and counts sorted by the number of counts

```python
sortedByCount = sorted(hashtagsDict.items(), key=lambda x: (x[1],x[0]), reverse=True)

splitintoTwoLists = [list(t) for t in zip(*sortedByCount)]
hashtags = splitintoTwoLists[0]
counts = splitintoTwoLists[1]
```

Now for the plotting

```python
plt.bar(range(len(counts[0:20])), counts[0:20], width = 1)
plt.xticks([x+0.5 for x in range(len(counts[0:20]))], hashtags[0:20], rotation='vertical')
```

Now, we also want to aggregate over time, so that we can understand how popular hashtags develop over time. To do this, we'll need to define another function for aggregation.  

```python
#This works if we don't care about time stuff. 
def ReadAndAggregateIntervals(linesAtATime,filepath):
    i = 0
    data = []
    hashtagsDict = {}
    #Create a named tuple that we'll be using as our dictionary key
    HashtagKey = namedtuple("HashtagKey", ["DoM", "Hour", "Text"])
    
    #Open and read in the file
    with open(filename) as cleanedTweets:
        for tweet in cleanedTweets:
            i += 1
            jsonline = json.loads(tweet)
            data.append(jsonline)
            #aggregate once we've read in the appropriate number of liens
            if (i % linesAtATime == 0):
                print i

                #Recode some of the data
                df = pd.DataFrame(data=data)
                df['DoM'] = df['created_at'].apply(lambda x: str(x)[8:10:])
                df['Hour'] = df['created_at'].apply(lambda x: int(str(x)[11:13:]))
                
                #loop over all the tweets
                for index,tweet in df.iterrows():
                    #Loop over the hashtags in the tweet
                    for hashtag in tweet['entities_hashtags_text']:
                        hashtagsDict[HashtagKey(
                                tweet["DoM"], 
                                tweet["Hour"], 
                                hashtag.upper())] = hashtagsDict.get((tweet["DoM"],
                                                             tweet["Hour"],
                                                             hashtag.upper()), 0) + 1
                #reset the data
                data = []
    #return the aggregation
    return hashtagsDict
```

Now, let's create a plot

```python
aggDict = ReadAndAggregateIntervals(100000,filename)
```

Now, let's create a plot of the most popular hashtags, allowing the user to input time and the number of hashtags they'd like to visualize

```python
def plotHashtags(DoM, Hour,nHashTags=20):
    #Filter the hashtags, so that we only get the tweets of the appropriate day of month and hour
    timeHashtags = {key: value for key,value in aggDict.items() if (key.DoM == str(DoM) and key.Hour == Hour)}

    #Sort the hashtags by count
    sortedByCount = sorted(timeHashtags.items(), key=lambda x: (x[1],x[0].Text), reverse=True)

    #Split them up into hashtags and counts
    dictList = [list(t) for t in zip(*sortedByCount)]
    hashtags = [x.Text for x in dictList[0]]
    counts = dictList[1]
    
    #Plot
    plt.title('Hashtag Distribution (Dec {0} at {1})'.format(DoM, Hour))
    plt.ylabel('Number of Tweets')
    plt.bar(range(len(counts[0:nHashTags])), counts[0:nHashTags], width = 1)
    plt.xticks([x+0.5 for x in range(len(counts[0:nHashTags]))], hashtags[0:nHashTags], rotation='vertical')
```

Now, we're going to create the sliders!

```python
DoM = widgets.IntSlider(min=10, max=27, value=15)
Hour = widgets.IntSlider(min=0, max=23, value=10)
nHashTags = widgets.IntSlider(min=0, max=40, value=50)
widgets.interact(plotHashtags, DoM=DoM, Hour=Hour, nHashTags=nHashTags)
```

```python
def aggregateHashtags(df, hashtagsDict):
    #Create a named tuple that we'll be using as our dictionary key
    HashtagKey = namedtuple("HashtagKey", ["DoM", "Hour", "Text"])
               
    df['DoM'] = df['created_at'].apply(lambda x: str(x)[8:10:])
    df['Hour'] = df['created_at'].apply(lambda x: int(str(x)[11:13:]))

    #loop over all the tweets
    for index,tweet in df.iterrows():
        #Loop over the hashtags in the tweet
        for hashtag in tweet['entities_hashtags_text']:
            hashtagsDict[HashtagKey(
                    tweet["DoM"], 
                    tweet["Hour"], 
                    hashtag.upper())] = hashtagsDict.get((tweet["DoM"],
                                                 tweet["Hour"],
                                                 hashtag.upper()), 0) + 1
            
    return hashtagsDict
    
```
